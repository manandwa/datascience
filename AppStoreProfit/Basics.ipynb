The following function explore_data takes in the following parameters and uses them to print rows in a easy to read way
1. dataset: defined as a list of lists
2. start: An integer that is the starting index of a data slice
3. end: An integer that is the ending index of a data slice
4. rows_and_columns: a Boolean variable is default to false

This function will loop through the defined data slice printing a row and adding a newline using the command 'print(\n)'
If the rows_and_columns variable is set to True (default is false) then it will print the number of rows and columns

The data should not have a header row as this will add an additional row and print the wrong number of rows

```python
def explore_data(dataset, start, end, rows_and_columns=False):
    dataset_slice = dataset[start:end]
    for row in dataset_slice:
        print(row)
        print('\n') # adds a new (empy line) after each row
        
        if rows_and_columns:
            print('Number of rows: ', len(dataset))
            print('Number of columns: ', len(dataset[0]))
```

```python
from csv import reader

def import_data(filename):
    file = open(filename)
    csvfile = reader(file)
    dataset = list(csvfile)
    return dataset
```

```python
# counting the number of duplicates
duplicate_apps = []
unique_apps = []
# if a an app name appears in the unique app list as we cycle through then it is a duplicate
for app in android:
    name = app[0]
    if name in unique_apps:
        duplicate_apps.append(name)
    else:
        unique_apps.append(name)

print('Number of duplicate apps: ', len(duplicate_apps))
```

Duplicates are not going to be randomly removed.  They will be removed by looking at the row with the highest number of reviews and removing the other duplicate rows

```python
reviews_max = {}
for app in android:
    name = app[0]
    n_reviews = float(app[3])
    if name in reviews_max and reviews_max[name] < n_reviews:
        reviews_max[name] = n_reviews
    
    elif name not in reviews_max:
        reviews_max[name] = n_reviews

        
    
```

```python
def is_english(string):
    non_english = 0
    for character in string:
        if ord(character) > 127:
            non_english+=1
    
    if non_english > 3:
        return False
    else:
        return True
```

Columns that we can use to genrate frequency tables are the following: Genre and reviews

We will use the display table function display_table() to display the frequency table of the columns that we want

```python
def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)

    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
```

We will now create the frequency table function freq_table() to generate the frequency table

```python
def freq_table(dataset, index):
    table = {}
    total = 0
    
    for row in dataset:
        total += 1
        column = row[index]
        if column in table:
                table[column] += 1
        else:
            table[column] = 1
            
    
    table_percentages = {}
    for key in table:
        percentage = (table[key] / total) * 100
        table_percentages[key] = percentage
    
    return table_percentages
```
