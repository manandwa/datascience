# Normalizing Text

```python
import re

def normalize_text(text):
    text = text.lower()
    text = re.sub('[^A-za-z0-9\s]','',text) # remove A-Z, a-z and 0-9 from the string
    return text
```

```python
jeopardy['clean_question'] = jeopardy['Question'].apply(normalize_text)
jeopardy['clean_answer'] = jeopardy['Answer'].apply(normalize_text)
jeopardy.head(10)
```

# Normalizing columns

```python
def normalize_values(text):
    text = re.sub('[^A-za-z0-9\s]','',text) # remove A-Z, a-z and 0-9 from the string
    try:
        text = int(text)
    except Exception:
        text = 0
    return text
```

```python
jeopardy['clean_value'] = jeopardy['Value'].apply(normalize_values)
jeopardy.head(10)
```

```python
jeopardy['Air Date'] = pd.to_datetime(jeopardy['Air Date'])
jeopardy.head()
```

# Answers in Questions

```python
# Create a function that counts how often the answer is deducible from the question
def count_matches(row):
    split_answer = row['clean_answer'].split(' ')
    split_question = row['clean_question'].split(' ')
    match_count = 0
    if 'the' in split_answer:
        split_answer.remove('the')
    if len(split_answer) == 0:
        return 0
    for item in split_answer:
        if item in split_question:
            match_count += 1
    matches = match_count / len(split_answer)
    return matches
```

```python
jeopardy['answer_in_question'] = jeopardy.apply(count_matches, axis=1)
```

```python
jeopardy.head()
```

```python
jeopardy['answer_in_question'].mean()
```

The answer will occur in the question about 6% of the time which means that we can't guess the question by hearing the answer.  We will need to study (which is what Ken Jennings did)

# Recycled questions

```python
question_overlap = []
terms_used = set()

jeopardy = jeopardy.sort_values('Air Date')
```

```python
for i, row in jeopardy.iterrows():
    split_question = row['clean_question'].split(' ')
    split_question = [q for q in split_question if len(q) > 5] # if characters are greater than 5 keep
    match_count = 0
    for word in split_question:
        if word in terms_used:
            match_count += 1
    for word in split_question:
        terms_used.add(word)
    if len(split_question) > 0:
        match_count /= len(split_question)
    question_overlap.append(match_count)
jeopardy['question_overlap'] = question_overlap
jeopardy['question_overlap'].mean()
```

This means that roughly 70% of questions are recyled but these are only in terms of single words not phrases.  Plus this also looks at a small subset of questions.  This means that we do need to look into recycled questions some more.

# Low value vs high value questions

```python
def determine_value(row):
    if row['clean_value'] > 800:
        value = 1
    else:
        value = 0
    return value

jeopardy['high_value'] = jeopardy.apply(determine_value, axis=1)
jeopardy.head()
```

```python
def count_usage(term):
    low_count = 0
    high_count = 0
    
    for i, row in jeopardy.iterrows():
        split_question = row['clean_question'].split(' ')
        if term in split_question:
            if row['high_value'] == 1:
                high_count += 1
            else:
                low_count += 1
    return high_count, low_count

observed_expected = []
comparison_terms = list(terms_used)[:5] # first 5 characters
observed_expected = []
for term in comparison_terms:
    observed_expected.append(count_usage(term))

observed_expected
```

# Applying the chisquared test

```python
from scipy.stats import chisquare
import numpy as np

high_value_count = jeopardy[jeopardy['high_value'] == 1].shape[0] # get the count from this mask
low_value_count = jeopardy[jeopardy['high_value'] == 0].shape[0] # get the count from this mask
```

```python
high_value_count
```

```python
low_value_count
```

```python
chi_squared = []
```

```python
for observ in observed_expected:
    total = sum(observ)
    total_prop = total / jeopardy.shape[0]
    high_value_expected = total_prop * high_value_count
    low_value_expected = total_prop * low_value_count
    observed = np.array([observ[0],observ[1]])
    expected = np.array([high_value_expected,low_value_expected])
    chi_squared.append(chisquare(observed,expected))

chi_squared
```

# Chi squared results

There is not a significance in terms used in the sample between high value and low value rows.  Only 1 term had a frequency higher than 5 which makes the chi squared test not useful here.  It would be better to run the test with terms that have higher frequencies
